```python
for tweet in tweets:
  frase = str(tb(tweet.text))
  data_e_hora_completa = datetime.now()
  data_string = data_e_hora_completa.strftime('%Y-%m-%d %H:%M:%S')
  dados = {"tweet": frase, "horario": data_string}
  print(frase)
  producer.send(topico, value=dados)
  sleep(1)
```
